### Spliting Data

```python
# independent features
x = data.iloc[:,2:7].values
x.shape
```

###### We have deopped the 'Temp', 'Humidity', 'Oct-Dec', 'avgjune' and 'sub' variables as they do not show a clear relation in their respective scatter plots.

```python
# dependent features
y = data.iloc[:,10:].values
y.shape
```

```python
# Splitting data into train and test sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

```python
# Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
import pickle
pickle.dump(sc,open("transform.pkl","wb"))
```

## Model Building

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='entropy', random_state=0)
dtc.fit(x_train,y_train)
```

```python
y_pred_dtc = dtc.predict(x_test)
```

```python
# Accuracy score of decision tree
from sklearn.metrics import accuracy_score
acc_dtc = accuracy_score(y_test, y_pred_dtc)
acc_dtc
```

```python
#Confusion Matrix of decision tree
from sklearn.metrics import confusion_matrix
cm_dtc = confusion_matrix(y_test, y_pred_dtc)
cm_dtc
```

```python
import sklearn.metrics as metrics
fpr_dtc, tpr_dtc, threshold_dtc = metrics.roc_curve(y_test, y_pred_dtc)
roc_auc_dtc = metrics.auc(fpr_dtc,tpr_dtc)
```

```python
import matplotlib.pyplot as plt
plt.title('ROC_AUC Curve of Decision Tree Classification')
plt.plot(fpr_dtc,tpr_dtc,'b',label = 'auc = %0.2f'%roc_auc_dtc)
plt.plot([0,1],[0,1],'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.xlabel('tpr')
plt.ylabel('fpr')
plt.legend(loc = 'lower right')
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 10, criterion='entropy', random_state=0)
rfc.fit(x_train,y_train)
```

```python
y_pred_rfc = rfc.predict(x_test)
```

```python
# Accuracy score of random forest
acc_rfc = accuracy_score(y_test, y_pred_rfc)
acc_rfc
```

```python
#Confusion Matrix of random forest
cm_rfc = confusion_matrix(y_test, y_pred_rfc)
cm_rfc
```

```python
fpr_rfc, tpr_rfc, threshold_rfc = metrics.roc_curve(y_test, y_pred_rfc)
roc_auc_rfc = metrics.auc(fpr_rfc,tpr_rfc)
```

```python
plt.title('ROC_AUC Curve of Random Forest Classification')
plt.plot(fpr_rfc,tpr_rfc,'b',label = 'auc = %0.2f'%roc_auc_rfc)
plt.plot([0,1],[0,1],'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.xlabel('tpr')
plt.ylabel('fpr')
plt.legend(loc = 'lower right')
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski')
knn.fit(x_train,y_train)
```

```python
y_pred_knn = knn.predict(x_test)
```

```python
# Acccuracy score of KNN
acc_knn = accuracy_score(y_test, y_pred_knn)
acc_knn
```

```python
#Confusion Matrix of KNN
cm_knn = confusion_matrix(y_test, y_pred_knn)
cm_knn
```

```python
fpr_knn, tpr_knn, threshold_knn = metrics.roc_curve(y_test, y_pred_knn)
roc_auc_knn = metrics.auc(fpr_knn,tpr_knn)
```

```python
plt.title('ROC_AUC Curve using KNN')
plt.plot(fpr_knn,tpr_knn,'b',label = 'auc = %0.2f'%roc_auc_knn)
plt.plot([0,1],[0,1],'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.xlabel('tpr')
plt.ylabel('fpr')
plt.legend(loc = 'lower right')
```

### Xgboost

```python
import xgboost
xgb = xgboost.XGBClassifier()
xgb.fit(x_train,y_train)
```

```python
y_pred_xgb = xgb.predict(x_test)
```

```python
# Accuracy score of xgboost
acc_xgb = accuracy_score(y_test, y_pred_xgb)
acc_xgb
```

```python
#Confusion Matrix of xgboost
cm_xgb = confusion_matrix(y_test, y_pred_xgb)
cm_xgb
```

```python
fpr_xgb, tpr_xgb, threshold_xgb = metrics.roc_curve(y_test, y_pred_xgb)
roc_auc_xgb = metrics.auc(fpr_xgb,tpr_xgb)
```

```python
plt.title('ROC_AUC Curve of XGBoost')
plt.plot(fpr_xgb,tpr_xgb,'b',label = 'auc = %0.2f'%roc_auc_xgb)
plt.plot([0,1],[0,1],'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.xlabel('tpr')
plt.ylabel('fpr')
plt.legend(loc = 'lower right')
```
